```python
from sklearn import datasets
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
import numpy as np

iris = datasets.load_iris()
x = pd.DataFrame(iris.data)
x.columns = ["sepal_length","sepal_width","petal_length","petal_width"]
y = pd.DataFrame(iris.target)
y.columns = ["target"]
colormap = np.array(["red","lime","green"])
```

```python
plt.subplot(1,2,1)
plt.scatter(x.petal_length,x.petal_width,c=colormap[y.target],s=40)
plt.title("actual clustering")
```

```python
model1 = KMeans(n_clusters=3)
model1.fit(x)

plt.subplot(1,2,2)
plt.scatter(x.petal_length, x.petal_width, c=colormap[model1.labels_],s=40)
plt.title("EM clustering")
plt.show()
```

```python
model2 = GaussianMixture(n_components=3)
model2.fit(x)

plt.subplot(1,2,2)
plt.scatter(x.petal_length, x.petal_width, c=colormap[model2.predict(x)],s=40)
plt.title("KMeans")
plt.show()
```
